```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
from scipy.stats import linregress
from pprint import pprint

# Import API key
# from api_keys import weather_api_key
weather_api_key = '5d006fc5bbde33dcb64a32e4463fe548'


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#build url and variables to stor data
url = "http://api.openweathermap.org/data/2.5/weather?appid=" + weather_api_key +"&units=imperial"

record = 1

city_name=[]
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# print(f'{url}&q={cities[0]}')
# for loop to set up return info
for city in cities:
    try:
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")

        record = record + 1
                           
    except:
        print('City not found. Skipping')
    continue   


```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Build DataFrame
city_df = {"City": city_name,
          "Cloudiness": cloudiness,
          "Country": country,
          "Date": date,
          "Humidity": humidity,
          "Lat": lat,
          "Lng": lng,
          "Max Temp": max_temp,
          "Wind Speed": wind_speed}
city_df = pd.DataFrame(city_df)
city_df
```

```python
#Output the CSV
city_df.to_csv(output_data_file)
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#build linear regression
x_value = north_df["Lat"].astype(float)
y_value = north_df["Wind Speed"].astype(float)

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
regress_value = x_value * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.scatter(x_value, y_value, marker='o', s=10)

plt.plot(x_value, regress_value, "r-")
plt.annotate(line_eq,(6,25), fontsize=15, color="red")

plt.xlabel("Latitude")
plt.ylabel("Max Temp")
plt.title(f'Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
print(f"R squared: {rvalue}")

#print data
plt.savefig("output_data/Northern Hemisphere Max Temp vs Latitude.png")
```
